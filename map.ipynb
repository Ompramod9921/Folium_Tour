```python
fg=folium.FeatureGroup(name="tourist places")
fg.add_child(folium.GeoJson('india_states.json'))
map.add_child(fg)
map
```

```python
data = pd.read_csv("places.csv")
```

```python
def color(name):
    if name == "National-Park" :
        return 'green'
    elif name == "Monument":
        return 'red'
    elif name == "Adventure-Sport":
        return 'darkblue'
    else:
        return 'orange'
```

```python
lat=list(data["LAT"])
lon=list(data["LON"])
type=list(data["TYPE"])
name=list(data["NAME"])
location=list(data["LOCATION"])
website=list(data["website"])
picture=list(data["picture"])
for lt,ln,ty,nm,loc,ws,pic in zip(lat,lon,type,name,location,website,picture):
    fg.add_child(folium.Marker(location=[lt,ln],popup="<b>Name  : </b>"+nm+ "<br> <b>Location : </b> "+loc+"<br><b>Wikipidea link: </b><a href="+ws+">click here</a>"+"<br> <img src="+pic+" height=142 width=290>",icon=folium.Icon(color=color(ty))))
map
```

```python
map.save("map.html")
```
